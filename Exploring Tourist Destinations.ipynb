# Deskripsi Dataset

Dataset [Indonesia Tourism Destination](https://www.kaggle.com/datasets/aprabowo/indonesia-tourism-destination?resource=download&select=package_tourism.csv) yang tersedia di Kaggle berisi informasi mengenai berbagai destinasi wisata di Indonesia. Dataset ini dibuat oleh pengguna dengan nama aprabowo dan terakhir diperbarui pada tahun 2021. Dataset ini mencakup beberapa kota besar di Indonesia seperti Jakarta, Yogyakarta, Semarang, Bandung, dan Surabaya.

| Nama Kolom      | Deskripsi                                                                 |
|-----------------|--------------------------------------------------------------------------|
| User_Id         | Identitas unik dari pengguna yang memberikan rating atau ulasan.         |
| Place_Id        | Identitas unik dari destinasi wisata.                                    |
| Place_Ratings   | Nilai rating yang diberikan oleh pengguna untuk destinasi wisata tersebut.|
| Location        | Lokasi umum dari destinasi wisata (kota, wilayah, atau daerah tertentu).  |
| Age             | Umur pengguna yang memberikan rating atau ulasan.                        |
| Place_Name      | Nama dari destinasi wisata.                                              |
| Description     | Deskripsi singkat mengenai destinasi wisata.                             |
| Category        | Kategori dari destinasi wisata (alam, budaya, sejarah, hiburan, dll.).   |
| City            | Kota di mana destinasi wisata tersebut berada.                           |
| Price           | Biaya tiket masuk atau harga lainnya yang terkait dengan destinasi wisata.|
| Rating          | Rating keseluruhan dari destinasi wisata berdasarkan agregasi rating pengguna.|
| Time_Minutes    | Waktu yang diperlukan untuk mengunjungi atau menikmati destinasi wisata tersebut (dalam menit).|
| Coordinate      | Koordinat geografis dari lokasi destinasi wisata (latitude dan longitude).|
| Lat             | Latitude, koordinat geografis lintang dari destinasi wisata.             |
| Long            | Longitude, koordinat geografis bujur dari destinasi wisata.              |
| Unnamed: 11     | Kolom yang mungkin kosong atau tidak terdefinisi dalam dataset.          |
| Unnamed: 12     | Kolom lain yang mungkin kosong atau tidak terdefinisi dalam dataset.     |
| Package         | Informasi mengenai paket wisata yang tersedia untuk destinasi tersebut, jika ada.|
| Place_Tourism1  | Destinasi wisata lain yang terkait atau direkomendasikan bersama dengan destinasi utama.|
| Place_Tourism2  | Destinasi wisata terkait kedua.                                          |
| Place_Tourism3  | Destinasi wisata terkait ketiga.                                         |
| Place_Tourism4  | Destinasi wisata terkait keempat.                                        |

# Latar Belakang

Kondisi saat ini, sekitar 60% destinasi wisata di Indonesia masih memiliki kelemahan seperti kurangnya infrastruktur dan fasilitas pendukung yang memadai seperti akses transportasi, akomodasi, restoran, dan fasilitas rekreasi. Menurut data Kementerian Pariwisata dan Ekonomi Kreatif, hanya 40% destinasi wisata yang memiliki aksesibilitas yang baik. Selain itu, belum optimalnya promosi dan pemasaran destinasi wisata juga menjadi kendala dalam menarik lebih banyak wisatawan. Data menunjukkan bahwa pertumbuhan kunjungan wisatawan mancanegara ke Indonesia hanya 4,2% pada 2022, jauh di bawah negara-negara tetangga seperti Malaysia (23,6%) dan Singapura (14,8%).Menurut laporan Badan Pusat Statistik, sebanyak 17% kawasan wisata alam mengalami kerusakan lingkungan akibat aktivitas pariwisata yang tidak terkendali. Oleh karena itu, analisis destinasi wisata diperlukan untuk memberikan rekomendasi perencanaan pengembangan yang lebih baik, strategi promosi dan pemasaran yang tepat, serta upaya pelestarian lingkungan dan budaya lokal, serta pemberdayaan masyarakat lokal untuk meningkatkan pendapatan dan kesejahteraan mereka dari sektor pariwisata yang saat ini baru menyumbang 4,3% terhadap PDB nasional.<br>
Maka dari itu saya ingin melakukan evaluasi mendalam terhadap setiap destinasi wisata untuk mengidentifikasi potensi yang dapat ditingkatkan serta kelemahan yang perlu diperbaiki, serta dapat diidentifikasi strategi promosi yang lebih efektif untuk meningkatkan kunjungan wisatawan baik domestik maupun mancanegara.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Data Preparation

## Data Loading

```python
df_package = pd.read_csv('package_tourism.csv')
df_package.tail()
```

```python
df_rating = pd.read_csv('tourism_rating.csv')
df_rating.tail()
```

```python
df_place = pd.read_csv('tourism_with_id.csv')
df_place.tail()
```

```python
df_user = pd.read_csv('user.csv')
df_user.tail()
```

```python
# Lakukan join menggunakan metode join()
result = df_rating.join(df_user.set_index('User_Id'), on='User_Id').join(df_place.set_index('Place_Id'), on='Place_Id').join(df_package.set_index('City'), on='City')

result.head()
```

## Data Cleaning

```python
# Tampilkan info dataframe
result.info()
```

```python
#menghapus kolom yang mungkin tidak diperlakukan untuk kebutuhan analisis
result=result.drop(['Unnamed: 11', 'Unnamed: 12'], axis=1)
result.info()
```

```python
# Cek kolom yang memiliki nilai null
null_columns = result.columns[result.isnull().any()].to_list()
null_columns
```

```python
# Isi nilai null
for col in null_columns:
    if result[col].dtype in ['float64', 'int64']:  # Jika kolom numerik
        result[col].fillna(0, inplace=True)
    else:  # Jika kolom string
        result[col].fillna('NaN', inplace=True)

# Tampilkan hasil
result.info()
```

## Make Function

```python
def visualize_time_distribution(df, group_by_col, value_col, title, xlabel):
    # Lakukan pengelompokan berdasarkan kolom yang ditentukan dan hitung total waktu yang dihabiskan
    time_spent = df.groupby(group_by_col)[value_col].sum().reset_index()

    # Visualisasikan distribusi waktu yang dihabiskan
    plt.figure(figsize=(10, 6))
    bars = plt.bar(time_spent[group_by_col], time_spent[value_col], color='skyblue')

    # Menambahkan label pada setiap batang
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')  # Menambahkan nilai di atas batang

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Total Waktu (menit)')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
    
```

```python
def visualize_places_distribution(df, group_by_col, value_col, title, xlabel):
    # Mengelompokkan berdasarkan kolom yang ditentukan dan menghitung jumlah unik value_col
    grouped_data = df.groupby(group_by_col)[value_col].nunique().reset_index()
    
    # Mengganti nama kolom untuk kejelasan
    grouped_data.columns = [group_by_col, 'Number_of_Places']

    # Visualisasi jumlah tempat destinasi berdasarkan kolom yang ditentukan
    plt.figure(figsize=(12, 8))
    bars = plt.bar(grouped_data[group_by_col], grouped_data['Number_of_Places'], color='skyblue')

    # Menambahkan label pada setiap batang
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')  # Menambahkan nilai di atas batang

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Jumlah Tempat Destinasi')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
```

```python
def visualize_total_visits(df, group_by_col, title, xlabel):
    # Mengelompokkan berdasarkan kolom yang ditentukan dan menghitung total kunjungan (jumlah baris)
    total_visits = df.groupby(group_by_col).size().reset_index(name='Total_Visits')

    # Visualisasi total kunjungan berdasarkan kolom yang ditentukan
    plt.figure(figsize=(12, 8))
    bars = plt.bar(total_visits[group_by_col], total_visits['Total_Visits'], color='skyblue')

    # Menambahkan label pada setiap batang
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')  # Menambahkan nilai di atas batang

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Total Kunjungan')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
```

```python
def visualize_revenue(df, group_by_col, title, xlabel):
    # Menghitung revenue
    df['Revenue'] = df['Price'] * df['User_Id']
    
    # Mengelompokkan berdasarkan kolom yang ditentukan dan menjumlahkan revenue
    total_revenue = df.groupby(group_by_col)['Revenue'].sum().reset_index()

    # Visualisasi distribusi revenue berdasarkan kolom yang ditentukan
    plt.figure(figsize=(12, 8))
    bars = plt.bar(total_revenue[group_by_col], total_revenue['Revenue'], color='skyblue')

    # Menambahkan label pada setiap batang
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')  # Menambahkan nilai di atas batang

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Total Revenue')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
```
